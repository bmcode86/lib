```python
#import library
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import scipy.stats as sts
```

```python
#ref:https://networkx.org/documentation/networkx-1.7/reference/generated/networkx.drawing.nx_pylab.draw_networkx.html
#G is the graph
G=nx.dodecahedral_graph()
nx.draw(G, pos=nx.spring_layout(G)) #use spring layout
```

```python
#Node size dependent on the node degree on NetworkX
#use property nodelist, node_size

#approach 1
#d=dict(G.degree)
#nx.draw(G, nodelist=d.keys(), node_size=[v * 100 for v in d.values()])
#approach 2
d1=G.degree
d = [(d1[node]+1) * 20 for node in G.nodes()] #transform d


nx.draw(G, nodelist=G.nodes(), node_size=[(d[v] +1)*5 for v in G.nodes()])
plt.show()
```
